```python
df = pd.read_csv("./songs.csv")
df.head()
```

```python
df = df.select_dtypes(include=np.number)
df.head()
```

```python
# Spliting dataframe according to year (< 2010 = training, >= 2010 testing)
TRAIN = df[df['year'] < 2010]
TEST = df[df['year'] >= 2010]
```

```python
x_train = TRAIN.loc[:, TRAIN.columns != 'Top10']
y_train = TRAIN.loc[:, TRAIN.columns == 'Top10']

x_test = TEST.loc[:, TEST.columns != 'Top10']
y_test = TEST.loc[:, TEST.columns == 'Top10']
```

```python
# Checking if there are any missing values
print(x_train.isnull().sum().all())
print(x_test.isnull().sum().all())
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(x_train, y_train)
```
