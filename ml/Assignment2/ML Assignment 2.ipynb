```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv" , delimiter=",") 
```

```python
target = data["Survived"]
data.drop("Survived" , axis = 1 , inplace = True )
```

```python
print(data.columns)
print(target.name)
```

```python
# Mean
x = data.copy()
print("Mean of the Age Columns : " , x.Age.mean())
x.Age.fillna(x.Age.mean() , inplace=True)
print("Age Column Filled With Mean() where Nan Entries : " , x.Age.head(15))
x.describe() 
```

```python
# Median
y = data.copy()
print(y.Age.head(10))
print("Median of the Age Columns : " , y.Age.median())
y.Age.fillna(y.Age.median() , inplace=True)
print("Age Column Filled With Median() where Nan Entries : " , y.Age.head(15))
y.describe() 
```

```python
# Mode 
z = data.copy() 
print("Median of the Age Columns : " , z.Age.mode().iloc[0])
z.Age.fillna(z.Age.mode().iloc[0] , inplace=True)
print("Age Column Filled With Median() where Nan Entries : " , z.Age.head(15))
z.describe()
```

```python
print("Total NaN values for Column Sex : " ,z.Sex.isnull().sum())
print("Total NaN values for Column Cabin : " ,z.Cabin.isnull().sum())
print("Total NaN values for Column Embarked : ",z.Embarked.isnull().sum())
z.Cabin.fillna(z.Cabin.mode().iloc[0],inplace=True)
z.Embarked.fillna(z.Embarked.mode().iloc[0],inplace=True)
print("Total NaN values for Column Cabin After Processing : " ,z.Cabin.isnull().sum())
print("Total NaN values for Column Embarked After Processing: ",z.Embarked.isnull().sum())
```

```python
#OneHotEncoding of the Categorical Variables
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
# One Hot Encoding Sex Column
transformed = ohe.fit_transform(z[['Sex']])
print(transformed.toarray())
```

```python
print(ohe.categories_[0])
# Add the encoded array in the table
z[ohe.categories_[0]] = transformed.toarray()
# Now Drop the Sex Column from the table
z.drop("Sex",axis = 1,inplace=True)
```

```python
# One Hot Encoding Cabin Column
transformed_Cabin = ohe.fit_transform(z[['Cabin']])
print(transformed_Cabin.toarray())
print(ohe.categories_)
#Here the Number of Categories are very high and this can lead to very high memory consumption So we use label Encoding
```

```python
from sklearn import preprocessing
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'Country'. 
z['Cabin']= label_encoder.fit_transform(z['Cabin']) 
print(z.head())
```

```python
transformed_Embarked = ohe.fit_transform(z[['Embarked']])
ohe.categories_
# As the Categories are less we use One Hot Encoding
```

```python
# Adding the Above Categories in the Table
z[ohe.categories_[0]] = transformed_Embarked.toarray()
#Dropping the Embarked Category from the table
z.drop('Embarked',axis=1,inplace=True)
```

```python
# Preprocessed Data for Errors and Categorical Variables
features = z.copy()
features.drop(['Name','Ticket'] , axis=1 , inplace=True)
features
```

```python
# Perform Standardization without Sklearn Library
for i in features.columns :
    features[i] = (features[i] - features[i].mean()) / features[i].std()
```

```python
features
```

```python
features2 = z.copy()
features2.drop(['Name','Ticket'] , axis=1 , inplace=True)
features2
```

```python
# Performing the Normalization Without Sklearn 
for column in features2.columns:
    features2[column] = (features2[column] - features2[column].min()) / (features2[column].max() - features2[column].min()) 

features2
```

```python
#Now to perform Standardization using the Sklearn Library
from sklearn.preprocessing import StandardScaler
features3 = z.copy()
features3.drop(['Name','Ticket'] , axis=1 , inplace=True)
scaler = StandardScaler()
features3 = scaler.fit_transform(features3)
features3
```

```python
# Perform the Normalization using Sklearn library
features4 = z.copy()
features4.drop(['Name','Ticket'] , axis=1 , inplace=True)
from sklearn import preprocessing
features4 = preprocessing.normalize(features4)
```
