```python
# Importing needed libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler
```

```python
# Loading dataset from csv file
titanic = pd.read_csv("./train.csv")
```

```python
# Independent variables (since we have to predict whether a person survived or not based on other column data)
titanic.loc[:, titanic.columns != "Survived"].head(4)
```

```python
# Dependent variable
titanic.loc[:, titanic.columns == "Survived"].head(4)
```

```python
# Missing value presence shown
titanic["Cabin"][1: 6]
```

```python
# Handling missing values
titanic["Cabin"] = titanic["Cabin"].fillna(method='ffill')
titanic["Cabin"][1: 6]
```

```python
# Encoded categorical data
encoded_data = pd.get_dummies(titanic, columns = ["Sex", "Cabin", "Embarked"])
encoded_data.head(5)
```

```python
# Checking for more missing data
titanic.isnull().sum()
```

```python
# Showing age before filling it with mean
titanic[titanic["Age"].isnull()].head(1)
```

```python
# Filling age with mean of the column (Passenger with Id 6 has been assigned age = 35.0)
mean_value = titanic["Age"].mean()
titanic["Age"] = titanic["Age"].fillna(value = int(mean_value))
titanic[titanic["PassengerId"] == 6]
```

```python
# Checking if there is any NaN age Left
titanic["Age"].isnull().sum()
```
